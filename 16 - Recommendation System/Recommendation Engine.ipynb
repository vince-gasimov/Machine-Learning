# About Dataset

```python
# def analyze_missing_values(data):
#     missing_df = pd.DataFrame({
#         "count":
#         data.isnull().sum(),
#         "percentage":
#         round(data.isnull().mean() * 100, 2)
#     })
#     missing_df = missing_df[missing_df['count'] != 0].sort_values(
#         by='count', ascending=True)
#     return missing_df
```

```python
def analyze_missing_values(data):
    
    if data.isnull().sum().sum() == 0:
        print("There is no missing value in the dataset.")
    else:
        missing_df = pd.DataFrame({
        "count":
        data.isnull().sum(),
        "percentage":
        round(data.isnull().mean() * 100, 2)
    })
        
        missing_df = missing_df[missing_df['count'] != 0].sort_values(
        by='count', ascending=True)
    
        return missing_df
```

```python
analyze_missing_values(data)
```

```python
data = data[~np.isnan(data["gross"]) & ~np.isnan(data["budget"])]
```

```python
analyze_missing_values(data)
```

```python
data = data[data.isnull().sum(axis=1) <= 2]
# data.isnull().sum(axis=1): This sums up the True values (which are treated as 1) for each row. 
# The axis=1 means "sum across the columns for each row." So, this part gives you the total number of
# missing values in each row.
```

```python
analyze_missing_values(data)
```

```python
# using mean for numeric columns
data['num_critic_for_reviews'].fillna(data['num_critic_for_reviews'].mean(),
                                      inplace=True)

data['duration'].fillna(data['duration'].mean(), inplace=True)

# using mode for categoric columns
data['language'].fillna(data['language'].mode()[0], inplace=True)

# For categorical data that cannot be filled with statistical values, such as player names, we use the term "Unknown Actor".
data['actor_2_name'].fillna('Unknown Actor', inplace=True)
data['actor_3_name'].fillna('Unknown Actor', inplace=True)
```

```python
analyze_missing_values(data)
```

```python
# Let's separate the movie genres into a different column
data['Moviegenres'] = data['genres'].str.split('|')
data['Genre1'] = data['Moviegenres'].apply(lambda x: x[0])

# Some movies have only one genre. In such cases, assign the same genre to 'Genre2' as well
data['Genre2'] = data['Moviegenres'].apply(lambda x: x[1]
                                           if len(x) > 1 else x[0])
data['Genre3'] = data['Moviegenres'].apply(lambda x: x[2]
                                           if len(x) > 2 else x[0])
data['Genre4'] = data['Moviegenres'].apply(lambda x: x[3]
                                           if len(x) > 3 else x[0])
```

## Analysis of Movies by a Specific Actor

```python
def query_actors(x):
    # Merge the movies where the specified actor is listed as the first, second, or third actor
    a = data[data['actor_1_name'] == x]
    b = data[data['actor_2_name'] == x]
    c = data[data['actor_3_name'] == x]
    y = pd.concat([a, b, c])

    y = y[[
        'movie_title',
        'budget',
        'gross',
        'title_year',
        'genres',
        'language',
        'imdb_score',
    ]]

    y = y.sort_values(by='imdb_score', ascending=False)
    return y
```

## Analysis of the Number of Reviews Evaluating Actors by Critics

```python
def critically_acclaimed_actors(m):
    frames = [
        data[data['actor_1_name'] == m], data[data['actor_2_name'] == m],
        data[data['actor_3_name'] == m]
    ]
    combined = pd.concat(frames)
    total_reviews = combined['num_critic_for_reviews'].sum()
    # check NaN values and conver to 0
    return int(total_reviews) if pd.notnull(total_reviews) else 0
```

# Recommending Movies Based on Actors

```python
def recommend_movies_on_actors(x):
    # Merge the movies where the specified actor is listed as the first, second, or third actor
    a = data[['movie_title', 'imdb_score']][data['actor_1_name'] == x]
    b = data[['movie_title', 'imdb_score']][data['actor_2_name'] == x]
    c = data[['movie_title', 'imdb_score']][data['actor_3_name'] == x]

    movies = pd.concat([a, b, c])

    # Sort by IMDb rating and return the top 15
    movies = movies.sort_values(by='imdb_score', ascending=False).head(15)

    return movies

```

**mlxtend (Machine Learning Extensions):**

- A Python library offering useful tools and algorithms for machine learning, data science, and data mining applications.

- mlxtend.preprocessing is the preprocessing module of mlxtend, containing various classes and functions for data preprocessing.

- This module facilitates a range of preprocessing tasks necessary to prepare datasets for machine learning algorithms, including feature scaling, converting categorical data to numerical data, and transforming datasets into binary formats.

- Operations include tasks like feature scaling, converting categorical data to numeric, and transforming data into binary formats.

**TransactionEncoder:**

- TransactionEncoder is a class used to convert items within a set of transactions (e.g., shopping cart transactions or movie genre lists in this case) into a binary format.

- It creates a binary matrix (True/False or 1/0) indicating the presence of an item in each transaction. This makes it easier to determine the presence of items for analyses like frequent itemset mining.

- In this case, TransactionEncoder takes lists of movie genres and produces a binary matrix indicating whether each genre is present in the movies.

- The resulting transformation prepares the dataset for genre analysis and modeling, where each column represents a movie genre, and each row shows whether that genre is present in the corresponding movie (1 or 0).

https://rasbt.github.io/mlxtend/user_guide/preprocessing/TransactionEncoder/




```python
data['genres']
```

```python
x = data['genres'].str.split('|')
```

```python
te = TransactionEncoder()
```

```python
x = te.fit_transform(x)
```

```python
x = pd.DataFrame(x, columns=te.columns_)
```

```python
x.head()
```

```python
genres = x.astype('int')

genres.head()
```

```python
genres.insert(0, 'movie_title', data['movie_title'])

genres.head()
```

```python
genres = genres.set_index('movie_title')

genres.head()
```

```python
def recommendation_genres(gen):
    gen = genres[gen]
    similar_genres = genres.corrwith(gen)
    similar_genres = similar_genres.sort_values(ascending=False)
    similar_genres = similar_genres.iloc[1:]
    return similar_genres.head(3)
```

```python
recommendation_genres('Drama')
```

```python
recommendation_genres('Action')
```

```python
recommendation_genres('Comedy')
```

# Recommending similar Movies

```python
x = genres.transpose()
x.head()
```

```python
def recommendation_movie(movie):
    movie = x[movie + '\xa0']
    similar_movies = x.corrwith(movie)
    similar_movies = similar_movies.sort_values(ascending=False)
    similar_movies = similar_movies.iloc[1:]
    return similar_movies.head(20)
```

```python
recommendation_movie('The Expendables')
```
